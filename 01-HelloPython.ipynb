## Basic data types

```python
x = 3
print(type(x)) # 데이터타입확인
print(x)       # x의 값을 출력
print(x + 1)   # 더하기
print(x * 2)   # 곱하기
print(x ** 2)  # 3^2
x += 1
print(x)  # x = x + 1 즉, x = 3 + 1
x *= 2
print(x)  # x = x * 2 즉, x = 4 * 2
```

```python
y = 2.5
print(type(y)) # 데이터타입확인
print(y, y + 1, y * 2, y ** 2) # y값, 1 더하기, 2 곱하기, 2.5^2
```

## For statement

range

```python
A = range(5) # A는 5개의 값을 가지고 있음
print(A)
```

```python
print(A[2]) # A의 인덱스 2번 출력
```

```python
for i in range(5): #반복문 for 변수명 in range(변수값):
    print(i, A[i])
```

```python
for i in range(3):
  for j in range(2):
    print('{} + {} = {}'.format(i, j, i+j))
```

## Operators

### + operator

```python
print((1, 2, 3) + (4, 5, 6))
print([1, 2, 3] + [4, 5, 6])
print("Hello" + " " + "World")
```

### * operator
The * operator produces a new tuple, list, or string that "repeats" the original content.

```python
print((1, 2, 3) * 3) # 3을 곱해서 출력이 아닌 3번 출력
print([1, 2, 3] * 3)
print("Hello "*3)
```

## Containers
Python includes several built-in container types: lists, dictionaries, sets, and tuples.

### Tuple
A simple immutable (변경할 수 없는, 불변의) ordered sequence of items

```python
# creating a tuple
months = ('January','February','March','April','May','June',\
'July','August','September','October','November','December')

print(months[0])
print("index of 7 ==> " , months[7])
```

```python
for i in range(12):
  print(months[i])
```

To print one by one

```python
# iterate through them:
for item in months:
    print (item)
```

```python
t = ('john', 32, (2,3,4,5), 'hello')
print(t)
print(t[2])
print(t[2][1]) # 2번에서 1번
print(t[:2]) # 2번 전 출력
print(t[2:]) # 2번 부터 출력

print(t[-1]) # 뒤에서 첫번째
print(t[-2])
```

```python
t[2] = 1
```

### List
Mutable (바꿀수 있는, 변경가능한) ordered sequence of items of mixed types

```python
li = ['hallym', 1, 3.141572, 'hello']
print(li)
li[1] = 45
print(li)
li.append('September') #append는 추가하다
print(li)
```

```python
v = []
for i in range(0,3):
    v.append(i)
    print(v)
```
